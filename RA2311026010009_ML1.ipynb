```python
import pandas as pd

# STEP 1: Load the CSV file
df = pd.read_csv("path_prediction_dataset.csv")

# STEP 2: Dataset overview
print("📄 DATASET STRUCTURE")
print("-" * 50)
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
print("Column Names:", list(df.columns))

# STEP 3: Info about data types and missing values
print("\n🔍 COLUMN DATA TYPES & MISSING VALUES")
print("-" * 50)
print(df.info())

# STEP 4: Descriptive statistics for numeric columns
print("\n📊 NUMERICAL STATISTICS")
print("-" * 50)
print(df.describe())

# STEP 5: Frequency count of key categorical columns
categorical_cols = ['aircraft_type', 'position_source', 'airspace_class', 'alert_level', 'on_ground']
print("\n📈 FREQUENCY COUNTS FOR CATEGORICAL COLUMNS")
print("-" * 50)
for col in categorical_cols:
    if col in df.columns:
        print(f"\n{col.upper()} value counts:")
        print(df[col].value_counts(dropna=False))

# STEP 6: Correlation matrix
print("\n🔗 CORRELATION MATRIX")
print("-" * 50)
print(df.corr(numeric_only=True))
```
